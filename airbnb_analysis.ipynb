TASK 1 : Installing all the dependecies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

TASK 2 : Loading the datasets

```python
data = pd.read_csv('datasets.csv', encoding_errors='ignore')
```

Task 4 : Data cleaning

```python
data.isnull().sum()
# Dropping missing values(Rows)
data.dropna(inplace=True)
#
data.isnull().sum()
```

```python
data.shape
```

```python
# Dealing with duplicated rows
data.duplicated().sum()   # tells the sum of duplicated row
# Deleting all duplicted rows
#data[data.duplicated()]   #showing all the duplicated rows
data.drop_duplicates(inplace=True)

data[data.duplicated()]
```

```python
#Type-Casting
#Changing data types
data.dtypes

data['id'] = data['id'].astype(object)
data['host_id'] = data['host_id'].astype(object) 
data.dtypes
```

Task 5 : Data Analysis

Univariate Analysis

```python
#identifying outliers
df = data[data['price'] < 1500]

sns.boxplot(data=df, x='price')
```

```python
#Price Distribution

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns
```

```python
 plt.figure(figsize=(6, 3))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

Feature Engineering

```python
#["Price per bed"]

df["price per bed"] = df['price']/df['beds']
df.head()
```

BI Variable Analysis

```python
df.columns
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# number of reviews and price relation
plt.figure(figsize=(10,8))
plt.title("Locality and Review Relation")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
#Geographical Distributions of airbnb
plt.figure(figsize=(10,7))
plt.title('Geographical Distributions of airbnb')
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.show()
```

```python
#HeatMap- correlation of one variable with other for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot=True)
```
